```python
def load_labels():
    memodb = MEMOGroupAff()
    labels_df = pd.read_csv(memodb.joint_labels_df_path)
    return labels_df, memodb.groups, memodb.sessions, memodb.emo_dims, memodb.annotators, memodb
```

```python
labels_df, groups, sessions, emo_dims, annotators, memodb = load_labels()
```

```python
labels_df["ArousalAvg"] = 0.0
labels_df["ValenceAvg"] = 0.0
```

```python
def get_annots_from_row(row, cols, emo_dim):
    req_cols = list(filter(lambda k: emo_dim+"_00" in k, cols)) 
    print(req_cols)
    annots = row[req_cols].dropna().values
    print(annots)
    return annots

for index, row in labels_df.iterrows():
    arous_avg = np.mean(get_annots_from_row(row, labels_df.columns, "Arousal"))
    valen_avg = np.mean(get_annots_from_row(row, labels_df.columns, "Valence"))
    
    labels_df.loc[index, "ArousalAvg"] = arous_avg
    labels_df.loc[index, "ValenceAvg"] = valen_avg
```
