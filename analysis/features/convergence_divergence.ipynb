# Convergence and Divergence Analyses

##### Utilities to Calculate and Plot correlations

```python
from utilities.plots import set_plot_params
import matplotlib.colors as mcolors
import matplotlib
import matplotlib.ticker as mtick
from scipy import stats
from matplotlib import ticker

from statsmodels.miscmodels.ordinal_model import OrderedModel
```

```python
def get_signal_category_colors(signal):
    color=[]
    btm_10 = np.percentile(signal, 10)
    top_10 = np.percentile(signal, 90)
    
    for i in range(len(signal)):
        if signal[i] <= btm_10:
            color.append('red')
        elif signal[i] > btm_10 and signal[i] < top_10:
            color.append('yellow')
        else:
            color.append('green')
    return color

def get_social_signal_and_affect(social_sig="pitch", agg_type="mean", video_sub_feat=""):
    
    arous_corr_list = []
    valen_corr_list = []
    arous_pval_list = []
    valen_pval_list = []
    
    arous_affect_list = []
    valen_affect_list = []
    
    mean_signal_list = []
    agg_signal_list = []
    
    for interaction in memoDB.interactions_df:
        arous_affect = getattr(interaction, "mean_arousal_gt")
        valen_affect = getattr(interaction, "mean_valence_gt")
        
        if "synchrony" in social_sig or "convergence" in social_sig:
            # synchrony_corrcoeff_pitch
            feature = social_sig.split("_")[-1]
            sub_feature = video_sub_feat
            feature_key = social_sig if sub_feature == "" else social_sig + "_" + sub_feature
            modality = interaction.get_ftype_modality(feature)
            feats_dict = interaction.audio_feats if modality == "audio" else interaction.video_feats
            mean_signal = feats_dict[feature_key]["mean"].values
            agg_signal  = feats_dict[feature_key][agg_type].values 
        else:
            modality = interaction.get_ftype_modality(social_sig)
            feats_dict = interaction.audio_feats if modality == "audio" else interaction.video_feats
        
            mean_signal = feats_dict["mean_"+ social_sig]
            agg_signal = feats_dict[agg_type+"_"+ social_sig] 
        
            if modality == "video":
                req_index = dutils.get_videofeats_index(social_sig, video_sub_feat)
                mean_signal = mean_signal[:, req_index] # Taking only the required video_sub_feat
                agg_signal  = agg_signal[:, req_index] # Taking only the required video_sub_feat
            
            if social_sig == "mfcc":
                mean_signal = mean_signal[:, 0] # Taking only the 1st MFCC coeff
                agg_signal  = agg_signal[:, 0] # Taking only the 1st MFCC coeff
            
        if arous_affect.shape[0] != mean_signal.shape[0]:
            arous_affect = arous_affect[:mean_signal.shape[0]]
            valen_affect = valen_affect[:mean_signal.shape[0]]

        arous_pcc_res = stats.kendalltau(arous_affect, agg_signal)#, method='auto', variant='c')
        valen_pcc_res = stats.kendalltau(valen_affect, agg_signal)#, method='auto', variant='c')
        
        arous_corr_list.append(arous_pcc_res.statistic)
        valen_corr_list.append(valen_pcc_res.statistic)
        
        arous_pval_list.append(arous_pcc_res.pvalue)
        valen_pval_list.append(valen_pcc_res.pvalue)
        
        arous_affect_list.extend(arous_affect.tolist())
        valen_affect_list.extend(valen_affect.tolist())
        
        mean_signal_list.extend(mean_signal.tolist())
        agg_signal_list.extend(agg_signal.tolist())

    arous_affect_list = np.array(arous_affect_list)
    valen_affect_list = np.array(valen_affect_list)
    
    mean_signal_list = np.array(mean_signal_list)
    agg_signal_list = np.array(agg_signal_list)
    
    return arous_affect_list, valen_affect_list, mean_signal_list, agg_signal_list,\
        np.mean(arous_corr_list), np.mean(valen_corr_list), np.mean(arous_pval_list), np.mean(valen_pval_list)

def fit_ordreg_and_calculate_r2(x, y, df):
    mod_prob = OrderedModel(x, y)
    res_prob = mod_prob.fit(method='bfgs', disp=True)    
    r2=0
    return mod_prob, res_prob, r2

def fit_poly2d_and_calculate_r2(x, y, df):
    # polynomial fit with degree df
    model = np.poly1d(np.polyfit(x, y, df))
    coeffs = np.round(model.coefficients, 4) 
    
    # fit values, and mean
    yhat = model(x)                         # or [p(z) for z in x]
    ybar = np.sum(y)/len(y)          # or sum(y)/len(y)
    ssreg = np.sum((yhat-ybar)**2)   # or sum([ (yihat - ybar)**2 for yihat in yhat])
    sstot = np.sum((y - ybar)**2)    # or sum([ (yi - ybar)**2 for yi in y])
    r2 = ssreg / sstot
    
    return model, coeffs, r2
    
import matplotlib as mpl

def plot_for_paper(social_sig="pitch", agg_type="std", plot=True, video_sub_feat="", dim="Arousal", ylabel="", save=True, log_scale=False, colorbar=False):
    # arous_affect_list, valen_affect_list, mean_signal_list, agg_signal_list, np.mean(total_arous_corr), np.mean(total_valen_corr)
    arous, valen, mean_signal, agg_signal, arous_corr, valen_corr, arous_pval, valen_pval = get_social_signal_and_affect(social_sig, agg_type, video_sub_feat)
    
    y_axis_signal    = mean_signal if "synchrony" in social_sig or "convergence" in social_sig \
                        else agg_signal
                        
    # if "convergence" in social_sig:
    #     y_axis_signal = y_axis_signal * -1
    
    if plot:
        
        width = 12 if not colorbar else 15
        fig = plt.figure(figsize=(width, 10))
        sns.set_style("whitegrid", {'axes.grid' : False})

        affect_list = arous if dim == "Arousal" else valen
        
        model, coeffs, r2 = fit_poly2d_and_calculate_r2(affect_list, y_axis_signal, df=2)
        # model, summ, r2 = fit_ordreg_and_calculate_r2(y_axis_signal, affect_list, df=2)                
        polyline = np.linspace(1, 9)
        
        if log_scale:
            plt.gca().set_facecolor(plt.get_cmap('viridis')(0))
            h = plt.hist2d(x=affect_list, y=y_axis_signal, cmap="viridis",
                           norm=mpl.colors.AsinhNorm(linear_width=2, vmin=.05), bins=30)
            
            if colorbar:            
                cbar = fig.colorbar(h[3], ticks=[9, 10, 15, 18, 20, 25])
                labels = np.array([0, 2, 5, 10, 20, 40, 90])#np.arange(0,100,30)
                loc    = labels + .5
                cbar.set_ticks(loc)
                cbar.set_ticklabels(labels)
        else:
            plt.gca().set_facecolor(plt.get_cmap('viridis')(0))
            plt.hist2d(x=affect_list, y=y_axis_signal, cmap="viridis", bins=30)
        
        # predicted = summ.model.predict(summ.params, exog=polyline)
        predicted = model(polyline)
        
        sns.lineplot(x=polyline, y=predicted, color='red', lw=6)

        plt.xlabel(dim)
        plt.ylabel(ylabel)
        plt.xlim((1, 9))
        # plt.ylim((-2,10))
        
        # plt.xticks([1,2,3,4,5,6,7,8,9])
        
        # cb = s.figure.colorbar() #.set_ticklabels([1, 5, 10, 15, 20, 28])
        # plt.subplots_adjust(left=0.27, right=0.73, top=0.8, bottom=0.2)  # shrink fig so cbar is visible
        
        # plt.hist2d(affect_list, y_axis_signal)

        set_plot_params(plt)
                       
        if save:
            fname = social_sig+ '_'+agg_type if video_sub_feat == "" else social_sig+ '_'+agg_type + '_' + video_sub_feat
            plt.savefig('plots_v4/'+dim.lower()+'_'+ fname + '.pdf', bbox_inches='tight')
        plt.show()

def calculate_convergence_divergence_correlations(social_sig="pitch", agg_type="std", plot=True, cat_color=False, video_sub_feat="", reg_curve=True):

    # arous_affect_list, valen_affect_list, mean_signal_list, agg_signal_list, np.mean(total_arous_corr), np.mean(total_valen_corr)
    arous, valen, mean_signal, agg_signal, arous_corr, valen_corr, arous_pval, valen_pval = get_social_signal_and_affect(social_sig, agg_type, video_sub_feat)
    
    y_axis_signal    = mean_signal if "synchrony" in social_sig or "convergence" in social_sig \
                        else agg_signal
    color_cat_signal = agg_signal if "synchrony" in social_sig or "convergence" in social_sig \
                        else mean_signal
    
    # if "convergence" in social_sig:
    #     y_axis_signal = y_axis_signal * -1
    
    if not cat_color:
        print("Arousal PCC: ", arous_corr, ". Valence PCC: ", valen_corr)
        print("Arousal PCC pval: ", arous_pval, ". Valence PCC pval: ", valen_pval)
        
        mine = MINE()
        mine.compute_score(arous, y_axis_signal)
        arous_mic = mine.mic()
        mine.compute_score(valen, y_axis_signal)
        valen_mic = mine.mic()
        print("Arousal MIC: ", arous_mic, ". Valence MIC: ", valen_mic)
        print()
        for i, affect_list in enumerate([arous, valen]):
            model, coeffs, r2 = fit_poly2d_and_calculate_r2(affect_list, y_axis_signal, df=2)                
            print("Regression Coefficients: ", "Arousal: " if i==0 else "Valence: ", str(coeffs[-3]) + " " + social_sig+"^2 + " + str(coeffs[-2]) + " " +social_sig + " + " + str(coeffs[-1]))
            print("R2 = ", r2)
        print()
        
    if plot:
        fig, axs = plt.subplots(1, 2, figsize=(10, 5))
        for i, affect_list in enumerate([arous, valen]):
            
            polyline = np.linspace(1.5, 9)
            axs[i].plot(polyline, model(polyline), color='red')
                    
            if cat_color:
                axs[i].scatter(affect_list, y_axis_signal, c=get_signal_category_colors(color_cat_signal))
            else:
                axs[i].hist2d(affect_list, y_axis_signal)
                

            axs[i].set(xlabel="Group Affect - " + "Arousal" if i==0 else "Valence", ylabel="Divergence/Convergence - " + agg_type +"_"+social_sig)

        # plt.xlabel("Valence")
        # plt.ylabel("Arousal")

        set_plot_params(plt)
        plt.show()
```
